```python
k = np.array([0,1,2,3,4,5,])
angle = k * 2 * np.pi / 6
x = np.cos(angle)
y = np.sin(angle)
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.grid(True)
plt.annotate("0",xy=(x[0],y[0]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))
plt.annotate("$e^{(1/3*\pi*i)}$",xy=(x[1],y[1]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))

plt.annotate("$e^{(2/3*\pi*i)}$",xy=(x[2],y[2]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))

plt.annotate("$e^{(\pi*i)}$",xy=(x[3],y[3]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))

plt.annotate("$e^{(-2/3*\pi*i)}$",xy=(x[4],y[4]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))

plt.annotate("$e^{(-1/3*\pi*i)}$",xy=(x[5],y[5]),
             xytext=(-20, +30),
             textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->'))
plt.show()
```

$1^{(1/6)} $

$ = 1 e^{0 + 2k\pi i},k \in {0,1,2,...} $

$ = 1 , k = 0 $

$ = e^{\frac{2}{6}\pi i } = cos(\frac{1}{3} \pi) + sin(\frac{1}{3}\pi)i ,k = 1 $

$ = e^{\frac{4}{6}\pi i } = cos(\frac{2}{3} \pi) + sin(\frac{2}{3}\pi)i ,k = 2 $

$ = e^{\frac{6}{6}\pi i } = cos(\pi) + sin(\pi)i = -1 ,k = 3 $

$ = e^{\frac{8}{6}\pi i } = cos(\frac{-2}{3} \pi) + sin(\frac{-2}{3}\pi)i ,k = 4 $

$ = e^{\frac{10}{6}\pi i } = cos(\frac{-1}{3} \pi) + sin(\frac{-1}{3}\pi)i ,k = 5 $
