```python
import numpy as np 
import matplotlib.pyplot as plt 
from collections import deque
from PIL import Image
```

```python
def mk_point(xs,ys):
    us = deque()
    vs = deque()
    for x in xs:
        for y in ys:
            u = np.sin(x) * np.cosh(y)  
            v = np.cos(x) * np.sinh(y)
            us.append(u)
            vs.append(v)

    us = list(us)
    vs = list(vs)
    return us,vs
```

```python
xs = np.linspace(-1*np.pi/2,np.pi/2,1000)
ys = np.array([0])
us,vs = mk_point(xs,ys)
plt.figure(figsize=(1,1))
plt.scatter(us,vs,s=1)
plt.show()
```

```python
xs = np.array([0])
ys = np.linspace(-10,10,1000)
us,vs = mk_point(xs,ys)
plt.figure(figsize=(1,1))
plt.scatter(us,vs,s=1)
plt.show()
```
